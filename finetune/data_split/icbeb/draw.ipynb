```python
import pandas as pd
import wfdb
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_path = "/home/tangsy/code/data/ICBEB/records500/402"
ecg = wfdb.rdsamp(data_path)
ecg = ecg[0].T
ecg.shape
```

```python
ecg=ecg[:,:5000]
leads = ['I', 'II', 'III', 'aVR', 'aVL', 'aVF', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6']
fs = 500  # 采样率 Hz
duration = ecg.shape[1] / fs  # 秒
time = np.linspace(0, duration, ecg.shape[1])

plt.figure(figsize=(13, 10))

for i in range(12):
    ax = plt.subplot(6, 2, i+1)
    
    # 红格背景：小格0.04s，大格0.2s（对应500Hz采样）
    ax.set_facecolor('#fff5f5')  # 背景淡粉色

    # 网格设置
    # 更大的格子：x小格0.1s，大格0.5s；y小格0.1mV，大格0.5mV
    # ax.set_xticks(np.arange(0, duration, 0.1), minor=True)
    # ax.set_xticks(np.arange(0, duration, 0.5))
    # ax.set_yticks(np.arange(-3, 3, 0.1), minor=True)
    # ax.set_yticks(np.arange(-3, 3, 0.5))
    # 
    # ax.grid(which='major', color='red', linewidth=0.6, alpha=0.3)
    # ax.grid(which='minor', color='red', linewidth=0.3, alpha=0.2)

    # 波形绘制（均值偏移处理）
    offset = np.mean(ecg[i])
    ax.plot(time, ecg[i] - offset, color='black', linewidth=1)

    ax.set_title(f'Lead {leads[i]}', fontsize=10, loc='left')

    ax.set_xlim(0, duration)
    ax.set_ylim(-1.5, 1.5)  # 显示区间根据ECG幅值适当调整
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```
